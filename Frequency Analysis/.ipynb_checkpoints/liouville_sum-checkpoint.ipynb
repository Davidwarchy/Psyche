```python
#make some imports needed
import math_functions
import numpy as np
import matplotlib.pyplot as plt
```

```python
#functions provided by https://wiki.math.ntnu.no/_media/ma3001/2014v/analytisktallteori/arithmeticalfunctions_tjerand
# check that n is in N \{0}.
def isValid (n):
    if (n <=0):
        return 0
    elif (n-int (n) != 0):
        return 0
    return 1

# Count the number of prime factors of n.
def numberOfFactors (n):
    counter = 0
    while (n >1):
        i = 2
        while (i <=n):
            if (n%i ==0):
                n /= i
                counter += 1
                break
            i += 1
    return counter

#The Liouville function .
# Liouville (1) = 1.
# Liouville (n) = ( -1) to the
# power of the sum of the divisors .
def Liouville (n):
    if not isValid (n):
        return 0
    if (n ==1):
        return 1
    else :
        return ( -1)**numberOfFactors (n)
```

```python
#create a line from 0 to N of numbers
N = 100000
n = np.arange(N)

#find the liouville of each number in N
sol = np.vectorize(Liouville)#so as to apply it to array
y = sol(n)#returns 1 where n has even number of prime factors and -1 where n has an odd number of prime factors
y = y.cumsum()#each value defines whether the mosts integers less than the integer corresponding to the index of the value in y have odd or even numbers of primes

fig,axs = plt.subplots(figsize=(20,8))
axs.grid(True)
axs.plot(n,y)
```

```python
#I got this idea from youtube https://www.youtube.com/watch?v=eQCUPQdi6DY --> wikipedia (Liouville Function)
#the idea is that the Lioville sum is scale invariant...
#now it would be interesting to find out the period of invariance, that is how much do we scale down/up a function such that it is invariant
```

```python
#lets make a guess and compare lsummation of the first 400 numbers with the summation function of the first 200 numbers
b = 4096
y2 = y[:b//2].repeat(2)
fig,axs = plt.subplots(figsize=(20,8))
axs.plot(n[:b],y[:b],n[:b],y2[:b])
axs.set_xlabel('larger scale')
axs.set_xlabel('liouville sum')
                       
def l2s(x):
    return x//2
def s2l(x):
    return x*2
                       
secax = axs.secondary_xaxis('top', functions=(l2s, s2l))
secax.set_xlabel('smaller scale')
plt.show()
```

```python
#I think what's most surprising about the above visual is the antisymmetry or symmetry about an invisible axis.
#It is obvious to the eye that there's a deep relationship between the two graphs
#performing a correlation doesn't show any significant values
corr = np.corrcoef(y[:b],y2[:b])
corr
#headings for columns and rows are y,y2,y,y2 respectively so that indices (0,0) show correlation of y with itself
#while index (0,1) shows the correlation of y2 with y. the values for correlations between the two diagrams are quite low
```

```python
#am hoping, like is always the case, a mathematician has done the job of helping us
#it's visible, the line about which the two curves oscillate.
#I think an easy way of finding that line is by finding the average and dividing by two. this may find the middle value
```

```python
y1 = y[:b]
y2 = y[:b//2].repeat(2)
yav = (y1+y2)/2
fig,axs = plt.subplots(figsize=(20,8))
axs.plot(n[:b],y1,n[:b],y2,n[:b],yav)
axs.set_xlabel('larger scale')
axs.set_xlabel('liouville sum')
                       
def l2s(x):
    return x//2
def s2l(x):
    return x*2
                       
secax = axs.secondary_xaxis('top', functions=(l2s, s2l))
secax.set_xlabel('smaller scale')
plt.show()
```
